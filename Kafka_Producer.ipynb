```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
producer = KafkaProducer(bootstrap_servers = ['Public IP Address'],
                        value_serializer = lambda x: dumps(x).encode('utf-8'))
```
